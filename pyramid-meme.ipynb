```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
x = [270, 70, 20]
y = ['Céu', "Lado ensolarado da pirâmide", "Lado sombreado da pirâmide"] 
mpl.rc('font', family = 'Comic Sans MS', size=16)
```

```python
plt.figure(figsize=(6,6))
plt.pie(x, startangle=-45, colors=['#0195da', 'gold', 'y'], radius = 0.8)
plt.legend(y, fontsize='large',bbox_to_anchor=(0.8,0.70), edgecolor="none")
plt.savefig("pyramid.png", dpi=100, bbox_inches='tight')
plt.show()
```
