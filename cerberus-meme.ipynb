```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
raio = 1 # raio dos circulos
distancia = raio/8 # distancia entre o centro do primeiro circulo e a linha do segundo circulo

plt.figure(figsize=(8,8))
plt.gca().add_patch(patches.Circle(xy=(raio,raio), radius=raio, fill=False))

plt.gca().add_patch(patches.Circle(xy=(raio + raio + distancia, raio), radius=raio, fill=False))

plt.gca().add_patch(patches.Circle(xy=(raio + (raio + distancia)/2, raio + (raio + distancia)*np.sqrt(3)/2), 
                                   radius=raio, fill=False))

plt.annotate("Cerberus", (raio + (raio + distancia)/2, raio + (raio + distancia)*np.sqrt(3)/2), 
                         xytext=(raio + (raio + distancia)/2, raio + (1/3)*(raio + distancia)*np.sqrt(3)/2), ha='center',
             fontsize=16
             )

plt.annotate("dog", (raio + (raio + distancia)/2, raio + (raio + distancia)*np.sqrt(3)/2), 
                         xytext=(raio + (raio + distancia)/2, 2*raio + (2/3)*(raio + distancia)*np.sqrt(3)/2), ha='center',
             fontsize=30
             )

plt.annotate("dog", fontsize=30, xy=(3*raio/5, raio), xytext=(3*raio/5, raio*1/3), ha='center')

plt.annotate("dog", fontsize=30, xy=(3*raio/5 + raio + raio, raio), 
             xytext=(3*raio/5 + raio + raio, raio*1/3), ha='center')

plt.axis(False)
plt.axis('equal')
# plt.savefig("cerberus.png", dpi=100, bbox_inches="tight")
plt.show()
```
